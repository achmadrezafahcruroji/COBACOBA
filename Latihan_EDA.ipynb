```python
def ngilangin(data):
  data.dropna(axis=0, thresh=4, inplace=True)
  data.dropna(axis=1, how='all', inplace=True)
  return data
```

```python
a = ngilangin(dfJ)
b = ngilangin(dfD)
c = ngilangin(dfT)
```

```python
df = pd.concat([a, b, c], ignore_index=True)
df
```

```python
#duplicate backward
df.bfill(inplace=True)
df.isnull().sum()
```
